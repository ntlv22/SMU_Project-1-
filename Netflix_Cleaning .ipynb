```python
import pandas as pd
```

```python
#read the files

file1 = "Resources/tv_shows.csv"
file2 = "Resources/MoviesOnStreamingPlatforms_updated.csv"

df = pd.read_csv(file1)
df1 = pd.read_csv(file2)
```

# TvShow

df.head()

```python
df.shape
```

```python
df.columns
```

```python
#drop columns
df_Tvshow = df.loc[:, ['Title', 'Year', 'Age', 'IMDb', 'Netflix','Hulu','Prime Video','Disney+','type']]
df_Tvshow.head()
```

```python
#convert "all" age into minimum age
df_Tvshow['Age'] = df_Tvshow['Age'].str.replace('all', '7+')
```

```python
#drop '+' on age column
df_Tvshow['Age'] = df_Tvshow['Age'].str.replace('+', '')
df_Tvshow.head()
```

```python
#use pd.to_numeric() method to convert the datatype of the age column
df_Tvshow['Age'] = pd.to_numeric(df_Tvshow['Age'])
```

```python
df_Tvshow.info()
```

```python
df_Tvshow.loc[pd.isnull(df_Tvshow.Age)].head(10)
```

```python
df_Tvshow.describe()
```
